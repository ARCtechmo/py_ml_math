```python
# Methods to Test for normality
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Data
# Is this data normaly distributed?
scores = np.array([88, 67, 64, 76, 86, 85, 82, 39, 75, 34, 90, 63, 89, 90, 84, 81, 96, 100, 70, 96])
```

```python
# Shapiro-Wilk Test for Normality
shapiro_test = stats.shapiro(scores)
print(f"Shapiro-Wilk Test statistic: {shapiro_test.statistic}, p-value: {shapiro_test.pvalue}")
if shapiro_test.pvalue > 0.05:
    print("Fail to reject the null hypothesis: Data is normally distributed.")
else:
    print("Reject the null hypothesis: Data is not normally distributed.")
```

```python
# Normal Probability plot (Q-Q Plot)
stats.probplot(scores, dist="norm", plot=plt)
plt.title("Normal Q-Q Plot")
plt.show()
```
