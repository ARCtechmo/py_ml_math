```python
import numpy as np
import sympy as sym
import pandas as pd
import matplotlib.pyplot as plt
import math
import statsmodels.api as sm
import seaborn as sns
import scipy.stats as stats
```

```python
# plotting
x = 3
y = 5
plt.plot(x,y,'bo') # hold shift + tab within the parenthesis to see all options
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()
plt.show()
```

```python
# basic plotting
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]
for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o', label='point %s' %(i+1))
plt.grid()
plt.legend()
plt.show()
```

```python
# plot a polynomial
x = sym.symbols('x')
y = x**2 - 3*x
for i in range(-11,12):
    plt.plot(i,y.subs(x,i),'o')
plt.grid()
plt.show()
```

```python
# graph a line
p1 = [-3,-1]
p2 = [4,4]

# [x1, y1], [x2, y2], color, line width
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color=[.4,.2,.7], linewidth=5)
plt.axis('square') # hold shift + tab within the parenthesis to see all options
plt.axis([-6,6, -6,6])
plt.grid()
plt.show()
```

```python
# graph a square root function y = sqrt(x)

# Define x values (non-negative only)
x = np.linspace(0, 20, 100)
y = np.sqrt(x)

# Plot the function
plt.plot(x, y, label=r"$y = \sqrt{x}$")

# Set the axis to be square
plt.axis('square')
plt.axis([-0, 20, 0, 8])
plt.grid(True)
plt.legend()
plt.title("Graph of $y = \sqrt{x}$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# slope intercept form
x = [-5,5]
m = 2
b = 1

# option 1: use numpy
y = m*np.array(x) + b

# option 2: use a for loop
# y = [340, 2345]
# for i in range(0,len(x)):
#     y[i] = m*x[i] + b


plt.plot(x,y,label='y=%sx+%s' %(m,b))
plt.axis('square')
plt.grid
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0], 'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()
plt.show()
```

```python
# graph a rational function 
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i]**2
plt.plot(x,y,'s-')
plt.show()
```

```python
# graph a rational function
x = np.linspace(-3,4,10)
y1 = 2 + np.sqrt(abs(x))
y2 = 2 + abs(x)**(1/3)
plt.plot(x,y1,'m-')
plt.plot(x,y2, 'k-')
plt.show()
```
