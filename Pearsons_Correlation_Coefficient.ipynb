```python
# Pearson's Correlation Coefficient (denoted as ùëü)
# measures the linear relationship between two variables. 
# ranges from -1 to 1
# r = 1: perfect positive reltionship
# r = -1: perfect negative relationship
# r = 0: no relationship
```

```python
## It is often very helful to apply visualtization techniques to better understand the correlation coefficients
# Steps to Create Visualizations
# 1) Generate datasets with different types of correlations.
# 2) Calculate Pearson's correlation coefficient for each dataset.
# 3) Visualize the datasets using scatter plots.
# 4) Annotate the plots with the correlation coefficient.
```

```python
# sidenote on the NumPy method 
# np.linspace is a numpy function used to create an array of evenly spaced values over a specific range
# np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

# Generate 10 evenly spaced numbers from 1 to 10
array = np.linspace(0,1,10)
print(array)


# start=0: The sequence starts at 0.
# stop=1: The sequence ends at 1.
# num=10: Generate 10 evenly spaced values.
# endpoint=True: Include the endpoint 1.
# retstep=False: Do not return the step size.
# dtype=None: Default to the data type of the input values.

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate datasets with specified correlations
np.random.seed(0)
n = 100

# Perfect positive correlation (r = 1)
x1 = np.linspace(0, 10, n)
y1 = x1

# Perfect negative correlation (r = -1)
x2 = np.linspace(0, 10, n)
y2 = -x2

# High positive correlation (r = 0.85)
x3 = np.linspace(0, 10, n)
y3 = 0.85 * x3 + np.random.normal(0, 1, n)

# High negative correlation (r = -0.85)
x4 = np.linspace(0, 10, n)
y4 = -0.85 * x4 + np.random.normal(0, 1, n)

# Moderate positive correlation (r = 0.5)
x5 = np.linspace(0, 10, n)
y5 = 0.5 * x5 + np.random.normal(0, 1, n)

# Moderate negative correlation (r = -0.5)
x6 = np.linspace(0, 10, n)
y6 = -0.5 * x6 + np.random.normal(0, 1, n)

# Low positive correlation (r = 0.25)
x7 = np.linspace(0, 10, n)
y7 = 0.25 * x7 + np.random.normal(0, 1, n)

# Low negative correlation (r = -0.25)
x8 = np.linspace(0, 10, n)
y8 = -0.25 * x8 + np.random.normal(0, 1, n)

# Very low positive correlation (r = 0.1)
x9 = np.linspace(0, 10, n)
y9 = 0.1 * x9 + np.random.normal(0, 1, n)

# Very low negative correlation (r = -0.1)
x10 = np.linspace(0, 10, n)
y10 = -0.1 * x10 + np.random.normal(0, 1, n)

# No correlation (r = 0)
x11 = np.random.rand(n)
y11 = np.random.rand(n)

# Plotting each dataset separately
datasets = [(x1, y1, 1), (x2, y2, -1), (x3, y3, 0.85), (x4, y4, -0.85), 
            (x5, y5, 0.5), (x6, y6, -0.5), (x7, y7, 0.25), (x8, y8, -0.25), 
            (x9, y9, 0.1), (x10, y10, -0.1), (x11, y11, 0)]
labels = ['r=1', 'r=-1', 'r=0.85', 'r=-0.85', 'r=0.5', 'r=-0.5', 'r=0.25', 'r=-0.25', 'r=0.1', 'r=-0.1', 'r=0']

fig, axs = plt.subplots(6, 2, figsize=(15, 30))
fig.delaxes(axs[5, 1])  # Remove the last empty subplot

for (x, y, r), label, ax in zip(datasets, labels, axs.flatten()):
    ax.scatter(x, y, alpha=0.6, label=label)
    # Plot line of best fit
    p = np.polyfit(x, y, 1)
    ax.plot(x, np.polyval(p, x), linestyle='dashed')
    ax.set_title(f'Scatter Plot with {label}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
