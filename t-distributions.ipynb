```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Sample sizes
sample_sizes = [5, 10, 30, 100]

# X-axis range for plotting
x = np.linspace(-4, 4, 1000)

# Plot standard normal distribution for comparison
plt.plot(x, norm.pdf(x), label='Standard Normal (Z)', color='black', linewidth=2)

# Plot t-distributions for various sample sizes
for n in sample_sizes:
    plt.plot(x, t.pdf(x, df=n-1), label=f't-distribution (n={n})')

# Plot settings
plt.title('Comparison of T-Distributions with Different Sample Sizes')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Adjust the y-axis limit to better highlight the differences
plt.figure(figsize=(10, 6))

# Plot standard normal distribution for comparison
plt.plot(x, norm.pdf(x), label='Standard Normal (Z)', color='black', linewidth=2)

# Plot t-distributions for various sample sizes with adjusted scaling
for n in sample_sizes:
    plt.plot(x, t.pdf(x, df=n-1), label=f't-distribution (n={n})')

# Plot settings
plt.title('Comparison of T-Distributions with Different Sample Sizes (Adjusted Scaling)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.ylim(0, 0.45)  # Adjusted y-axis limit to highlight differences
plt.legend()
plt.grid(True)
plt.show()
```
