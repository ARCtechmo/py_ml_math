```python
# frequency distributions
# exercise 2.20 - 2.25
import matplotlib.pyplot as plt

# initialize an empyt dictionary to store the frequeny of each letter
freq_dict = {}
letters = ['A','C','B','A','A','A','B','B','D','D']

# count the frequency of each letters
for letter in letters:
    if  letter in freq_dict:
        freq_dict[letter] +=1
    else:
        freq_dict[letter] = 1
        
# calculate frequency totals
total_freq = sum(freq_dict.values())

# Display frequency distribution 
print("Frequency Distribution:", freq_dict)


# Calculate and display relative frequencies
relative_freq_dict = {k: v / total_freq for k, v in freq_dict.items()}

# plot the bar chart
plt.bar(freq_dict.keys(), freq_dict.values())
print("Relative Frequencies:", relative_freq_dict)


# Plotting the bar chart for frequencies
plt.bar(freq_dict.keys(), freq_dict.values(), color='b', alpha=0.7, label='Frequency')
plt.bar(relative_freq_dict.keys(), relative_freq_dict.values(), color='r', alpha=0.5, label='Relative Frequency')

# add title and labels
plt.title('Frequency Distribution of Letters')
plt.xlabel('letters')
plt.ylabel('Frequency')
plt.legend()

# display the bar chart
plt.show()
```

```python
# frequency distributions
# exercise 2.32
import matplotlib.pyplot as plt

# Frequency data for different types of robberies
freq_dict = {
    'Street/highway': 127403,
    'Commercial house': 37885,
    'Gas/service station': 7009,
    'Convenience store': 14863,
    'Residence': 49361,
    'Bank': 5777,
    'Miscellaneous': 48878
}

# Calculate total frequency
total_freq = sum(freq_dict.values())

# Calculate and display relative frequencies, rounded to two decimal places
relative_freq_dict = {k: round(v / total_freq, 2) for k, v in freq_dict.items()}
print("Relative Frequencies:", relative_freq_dict)

# Draw a pie chart for relative frequencies
plt.figure()
plt.pie(relative_freq_dict.values(), labels=relative_freq_dict.keys(), autopct='%1.2f%%')
plt.title('Relative Frequency Distribution of Robbery Types')
plt.show()

# Draw a bar chart for frequencies
plt.figure()
plt.bar(freq_dict.keys(), freq_dict.values(), color='b', alpha=0.7)
plt.title('Frequency Distribution of Robbery Types')
plt.xlabel('Robbery Types')
plt.ylabel('Frequency')
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# frequency distributions
# exercise 2.38
import matplotlib.pyplot as plt

# Frequency data for different health statuses for two age groups
freq_19_64 = {'Excellent/very good': 119.7, 'Good': 48.4, 'Fair or poor': 20.2}
freq_65_over = {'Excellent/very good': 18.1, 'Good': 14.2, 'Fair or poor': 9.2}

# Calculate total frequencies for each age group
total_freq_19_64 = sum(freq_19_64.values())
total_freq_65_over = sum(freq_65_over.values())

# Calculate and display relative frequencies, rounded to two decimal places
relative_freq_19_64 = {k: round(v / total_freq_19_64, 2) for k, v in freq_19_64.items()}
relative_freq_65_over = {k: round(v / total_freq_65_over, 2) for k, v in freq_65_over.items()}

print("Relative Frequencies for age 19-64:", relative_freq_19_64)
print("Relative Frequencies for age 65 and over:", relative_freq_65_over)

# Draw pie charts for both age groups
plt.figure()
plt.pie(relative_freq_19_64.values(), labels=relative_freq_19_64.keys(), autopct='%1.2f%%')
plt.title('Relative Frequency Distribution of Health Status for Ages 19-64')
plt.show()

plt.figure()
plt.pie(relative_freq_65_over.values(), labels=relative_freq_65_over.keys(), autopct='%1.2f%%')
plt.title('Relative Frequency Distribution of Health Status for Ages 65 and Over')
plt.show()
```
