```python
# example 2.20 
# Stem and Leaf Plot
import matplotlib.pyplot as plt
from collections import defaultdict

# Given data
data = [70, 64, 99, 55, 64, 89, 87, 65, 62, 38, 67, 70, 60, 69, 78, 39, 75, 56, 71, 51, 99, 68, 95, 86, 57, 53, 47, 50, 55, 81, 80, 98, 51, 36, 63, 66, 85, 79, 83, 70]

# Create stem-and-leaf plot data
stem_leaf = defaultdict(list)
for num in data:
    stem, leaf = divmod(num, 10)
    stem_leaf[stem].append(leaf)
for stem in stem_leaf:
    stem_leaf[stem].sort()

# Create subplot for stem-and-leaf
plt.subplot(1, 2, 2)

# Plotting the stem-and-leaf plot
plt.axis([0, 4, min(data) // 10 * 10 - 5, max(data) // 10 * 10 + 5])
plt.yticks(range(min(data) // 10 * 10, max(data) // 10 * 10 + 10, 10))

for stem, leaves in sorted(stem_leaf.items()):
    plt.text(1, stem * 10, f"{stem} | {' '.join(map(str, leaves))}")

# Adding labels and title for stem-and-leaf
plt.xlabel('Leaf')
plt.ylabel('Stem')
plt.title('Stem-and-Leaf Plot')

# Show both plots
plt.tight_layout()
plt.show()
```

```python
# exercise 2.71
# histogram and frequency tables
import matplotlib.pyplot as plt

# Given data
data = [43.01, 40.39, 38.55, 38.01, 42.55, 41.21, 36.16, 35.68, 36.98, 30.28, 30.90, 39.06, 35.94, 25.91, 33.57, 31.44, 42.13, 34.26, 35.25, 33.02, 37.26, 31.30, 39.25, 34.93, 32.46]

# Calculate bin edges based on cutpoint grouping
first_cutpoint = 25
class_width = 3
max_value = max(data)
bin_edges = [first_cutpoint + i*class_width for i in range(0, int((max_value - first_cutpoint) / class_width) + 2)]

# Create frequency table
print("Cutpoint Range", "Count", "Relative Frequency", sep="\t")
total_data_points = len(data)
for i in range(len(bin_edges) - 1):
    count = sum(bin_edges[i] <= val < bin_edges[i+1] for val in data)
    rel_freq = count / total_data_points
    print(f"{bin_edges[i]} - under {bin_edges[i+1]}", count, f"{rel_freq:.2f}", sep="\t")

# Calculate weights for relative frequency
weights = [1/len(data)] * len(data)

# Plotting the histogram with relative frequencies
plt.hist(data, bins=bin_edges, edgecolor='black', align='left', weights=weights)
plt.xlabel('Data Range')
plt.ylabel('Relative Frequency')
plt.title('Histogram with Cutpoint Grouping')
plt.show()
```
