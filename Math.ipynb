```python
# exercise
x,y = sym.symbols('x y')
Fxy = (4+x)*(2-y)

xyset = range(0,3)

outmat = np.zeros([len(xyset),len(xyset)])
outmat

for i in xyset:
    for j in xyset:
        outmat[i,j] = Fxy.subs({x:i,y:j})
display( Math(sym.latex(sym.sympify(outmat))))   
```

```python
# bar plots
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# data sizes
m = 30 # rows
n = 6 # columns

# generate data
data = np.zeros((m,n))
for i in range(n):
    data[:,i] = 30*np.random.randn(m)*(2*i/(n-1)-1)**2 + (i+1)**2
    
# show the bars
fig,ax = plt.subplots(1,3,figsize=(8,2))
ax[0].bar(range(n),np.mean(data,axis=0))
ax[0].set_title('Bar plot')


# just the error bars
ax[1].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')
ax[1].set_title('Errorbar plot')

# both
ax[2].bar(range(n),np.mean(data,axis=0))
ax[2].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='.',linestyle='',color='k')
ax[2].set_title('Error+bar plot')

plt.show()
```

```python
# create a box and whisker plot of a Gaussian distribution
import matplotlib.pyplot as plt
import numpy as np

# create random data from a normal distribution
mean = 0
st_dev = 1
data = np.random.normal(mean, st_dev, 1000)

# create a vertical box and whisker plot
plt.subplot(2,2,1)
plt.boxplot(data)
plt.title('Vertical Box and Whisker Plot of a Gaussian Distribution')
plt.ylabel('Value')

# Create a horizontal box and whisker plot
plt.subplot(2, 1, 2)
plt.boxplot(data, vert=False)
plt.title('Horizontal Box and Whisker Plot of a Gaussian Distribution')
plt.xlabel('Value')

plt.tight_layout()
plt.show()
```

```python
# prime factorization
import sympy as sym
number = int(input("number: "))

# determine the prime factorization
# output is a dictionary key, value pair
fact_dict = sym.factorint(number)
print(fact_dict)
print(fact_dict.keys())
list(fact_dict.keys())


# prime number key length
print(len(list(fact_dict.keys())))
```

```python
# prime factorization
# loop through integers from 2 through 50
# determine if the integer is prime

for i in range(2,51):
    # determine the prime factorization
    # output is a dictionary key, value pair
    fact_dict = sym.factorint(i)
    if len(list(fact_dict.keys())) > 1:
        print(f'{i} is a composite number with prime factors {(list(fact_dict.keys()))}')
    else:
        print(f'{i} is a prime number.')
```

```python
# create a pie chart
# import libraries
import matplotlib.pyplot as plt
import numpy as np

## create data for the plot

nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1/2),np.log10(nbins-.01),totalN))


# prepare data for pie chart
uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))

for i in range(len(uniquenums)):
    data4pie[i] = sum(rawdata==uniquenums[i])
#     print(i)
    
# show the pie chart
plt.pie(data4pie,labels=100*data4pie/sum(data4pie))
plt.show()
```

```python
# plot lines
# import libraries
import matplotlib.pyplot as plt
import numpy as np

## create data for the plot

# number of data points
n = 1000

# generate log-normal distribution
data1 = np.exp( np.random.randn(n)/2 )
data2 = np.exp( np.random.randn(n)/10 )
data3 = np.exp( np.random.randn(n)/2 + 1 )

## plots of their histograms

# number of histogram bins
k = 20

plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)

plt.show()

# histogram discretization for the datasets
y1,x1 = np.histogram(data1,bins=k)
xx1 = (x1[0:-1] + x1[1:]) / 2
y1 = y1 / sum(y1) # convert to probability

y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[0:-1] + x2[1:]) / 2
y2 = y2 / sum(y2) # convert to probability

y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[0:-1] + x3[1:]) / 2
y3 = y3 / sum(y3) # convert to probability

# show the plots
plt.plot(xx1,y1,'s-',label='data1')
plt.plot(xx2,y2,'o-',label='data2')
plt.plot(xx3,y3,'^-',label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

```python
# plot lineaar y-axis
# plot a logarithmic y-axis
# y = e^x 
# y = x

# Step 1: Import necessary modules
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Create an array of x values
x = np.linspace(0, 10, 100)  # Generating 100 x values from -10 to 10

# Step 3: Calculate corresponding y values (y = e^x)
y_exp = np.exp(x)

# Step 4: Plot y = e^x
plt.plot(x, y_exp, label='y = e^x')  # Plotting y = e^x
plt.xlabel('x')                      # Label for x-axis
plt.ylabel('y')                      # Label for y-axis
plt.title('Graph of y = e^x')        # Title of the plot
plt.legend()                         # Display legend
plt.grid(True)                       # Display grid
plt.axhline(1, color='black', linewidth=0.5)  # Adding horizontal line at y = 1
plt.axvline(0, color='black', linewidth=0.5)  # Adding vertical line at x = 0

# Plot y = x
plt.plot(x, x, label='y = x', linestyle='dashed')  # Plotting y = x with dashed line

# Adjust y-axis limits for better visibility of y = x
plt.ylim(0, 10)  # Set y-axis limits from -10 to 10

plt.legend()   # Display legend for both y = e^x and y = x
plt.show()     # Display the plot
```

```python
# plot a logarithmic y-axis
# y = e^x 
# y = x

# Step 1: Import necessary modules
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Create an array of x values
x = np.linspace(0, 2, 100)  # Generating 100 x values

# Step 3: Calculate corresponding y values (y = e^x)
y_exp = np.exp(x)

# Step 4: Plot the points using Matplotlib with a logarithmic y-axis
plt.semilogy(x, y_exp, label='y = e^x')  # Plotting y = e^x with logarithmic y-axis

# Plotting y = x
plt.semilogy(x, x, label='y = x', linestyle='dashed')  # Plotting y = x with logarithmic y-axis

plt.xlabel('x')                      # Label for x-axis
plt.ylabel('y')                      # Label for y-axis (logarithmic scale)
plt.title('Graph of y = e^x and y = x')  # Title of the plot
plt.legend()                         # Display legend
plt.grid(True)                       # Display grid
plt.axhline(1, color='black', linewidth=0.5)  # Adding horizontal line at y = 1
plt.axvline(0, color='black', linewidth=0.5)  # Adding vertical line at x = 0
plt.show()                           # Display the plot
```
